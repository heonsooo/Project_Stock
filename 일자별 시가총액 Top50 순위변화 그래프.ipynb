```python
import numpy as np
import pandas as pd

import plotly
import matplotlib.pylab as plt
import plotly.graph_objs as go
# import plotly.io as pio
# pio.renderers.default = "notebook_connected"
```

```python
rank = pd.read_csv(r"시가총액(Top50).csv", index_col = 0)
# rank = rank.transpose()
rank = rank[rank.columns]
# rank
```

```python
fig = go.Figure()
for column in rank.columns:
    fig.add_trace(go.Scatter(x=rank.index, y=rank[column], name=column+'위'))

fig.update_yaxes(autorange="reversed")
fig.update_layout(
    title="일자별 시가총액 Top50 순위 변화",
    xaxis_title="일자",
    yaxis_title="순위",
    width=900,
    height=1200)
fig.show()
```
